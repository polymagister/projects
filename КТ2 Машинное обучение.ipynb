7. Строю кастомный трансформер на основе IQR

```python
from sklearn.base import BaseEstimator, TransformerMixin

class RareCategoryGrouper(BaseEstimator, TransformerMixin):
    def __init__(self, min_freq=10):
        self.min_freq = min_freq
        self.common_categories_ = None
    def fit(self, X, y=None):
        X_series = pd.Series(X)
        value_counts = X_series.value_counts()
        self.common_categories_ = set(value_counts[value_counts >= self.min_freq].index)
        return self

    def transform(self, X):
        X_series = pd.Series(X).copy()
        X_series[~X_series.isin(self.common_categories_)] = "Other"
        return X_series.values
```
